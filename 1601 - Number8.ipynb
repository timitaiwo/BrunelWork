__Generate Two Graphs over an Axis__

Two graphs would be drawn over the same axis that follows the equations;

i. y = Bcos(Cπx)\
ii. y=Bcos(Aπx)$e^{-x}$

where A, B and C are given numbers

```python
# Import numpy, pandas, matplotlib.plt and seaborn
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
#sns.set_theme(style="darkgrid")     # Set the format for diagrams
```

```python
# Set the values for A, B and C
a, b, c = [21, 11, 4]

# generate the vales for x
x = np.array([i for i in np.arange(0, 2.001, 0.001)])

# set the equation for both curves
y1 = b * np.cos(c * np.pi * x)
y2 = b * np.cos(a * np.pi * x) * np.exp(-x)
x_y_y2 = pd.DataFrame({'x':x, 'y = 11cos(4πx)':y1, r'y=11cos(21πx)$e^{-x}$':y2})
```

```python
# Plot the graphs
fig, ax = plt.subplots(figsize=(25,15))

# Plot the curves of y1 and y2 with respect to x
ax.plot(x, y1, "k", label="y = 11cos(4πx)")
ax.plot(x, y2, "k--", label=r"y=11cos(21πx)$e^{-x}$")

# Set the limits for the x-axis and y-axis
ax.set_xlim(0, 2)
ax.set_ylim(-13.5, 11.5)

# Label the graph
font_prop = {"size":20}
ax.set_xlabel("X Axis", **font_prop)
ax.set_ylabel("Y Axis", **font_prop)
plt.legend(loc="lower left", fontsize=20)
ax.set_title("1601 Coursework, Number 8", **font_prop)

# Save the image
plt.savefig("Number8.jpg")
```
