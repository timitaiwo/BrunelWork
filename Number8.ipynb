```python
# Import numpy, pandas, matplotlib.plt and seaborn
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_theme(style="darkgrid")     # Set the format for diagrams
```

```python
# Set the values for A, B and C
a, b, c = [21, 11, 4]

# generate the vales for x
x = np.array([i for i in np.arange(0, 2.001, 0.001)])

# set the equation for both curves
y1 = b * np.cos(c * np.pi * x)
y2 = b * np.cos(a * np.pi * x) * np.exp(-x)
x_y_y2 = pd.DataFrame({'x':x, 'y = 11cos(4πx)':y1, r'y=11cos(21πx)$e^{-x}$':y2})
```

```python
# Plot the graphs
fig, ax = plt.subplots(figsize=(25,10))

# Plot the curves of y1 and y2 with respect to x
ax.plot(x, y1, "k")
ax.plot(x, y2, "k--")

# Set the limits for the x-axis and y-axis
ax.set_xlim(0, 2)
ax.set_ylim(-13.5, 11.5)

# Label the graph
ax.set_xlabel("X Axis", size=20)
ax.set_ylabel("Y Axis", size=20)
ax.set_title("1601 Coursework, Number 8", size=20)
```
